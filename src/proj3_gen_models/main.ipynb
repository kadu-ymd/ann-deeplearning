```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img1 = mpimg.imread('img/SD1.5_00055_.png')
img2 = mpimg.imread('img/SD1.5_00056_.png')
img3 = mpimg.imread('img/SD1.5_00057_.png')
img4 = mpimg.imread('img/SD1.5_00058_.png')
img5 = mpimg.imread('img/SD1.5_00059_.png')
img6 = mpimg.imread('img/SD1.5_00060_.png')
```

```python
fig, axs = plt.subplots(3, 3, figsize=(8, 6))

# Mostra cada imagem em um subplot
axs[0, 0].imshow(img1)
axs[0, 0].set_title('Imagem 1')

axs[0, 1].imshow(img2)
axs[0, 1].set_title('Imagem 2')

axs[0, 2].imshow(img3)
axs[0, 2].set_title('Imagem 3')

axs[1, 0].imshow(img4)
axs[1, 0].set_title('Imagem 4')

axs[1, 1].imshow(img5)
axs[1, 1].set_title('Imagem 5')

axs[1, 2].imshow(img6)
axs[1, 2].set_title('Imagem 6')

for ax in axs.flat:
    ax.axis('off')

# plt.tight_layout()
plt.show()
```
